```python

# EXERCISE 72
# A string is a palindrome if it is identical forward and backward.
# For example, "anna", "civic", "level", and "hannah" are all examples
# of palindromic words.
# Write a program that reads a string from the user and uses a loop
# to determine whether or not it is a palindrome. Display the result,
# including a meaningful

word = input("Enter a word or string to check: ").lower()
is_palindrome = True
length = len(word)
i = 0
while i < (length // 2) and is_palindrome:
    if word[i] != word[length - 1 - i]:
        is_palindrome = False

    i += 1
if is_palindrome:
    print(f"{word} is a palindrome.")
else:
    print(f"{word} is not a palindrome.")
```

```python

# Number 75
# The greatest common divisor of two positive integers, n and m, is the
# largest number, d, which divides evenly into both n and m.
# 1. Initialize d to the smaller of m and n.
# 2. While d does not evenly divide m or d does not evenly divide n do:
#      Decrease the value of d by 1
# 3. Report d as the greatest common divisor of n

n = int(input("Enter the first positive integer (n): "))
m = int(input("Enter the second positive integer (m): "))
if m < n:
    d = m
else:
    d = n
while m % d != 0 or n % d != 0:
    d -= 1
print(f"The greatest common divisor of {n} and {m} is: {d}")
```

```python

# EXERCISE 70
# One of the first known examples of encryption was used by Julius Caesar.
# This program shifts each letter in a message by a specified amount,
# using a while-loop instead of a for-loop

LOWER = "abcdefghijklmnopqrstuvwxyz"
UPPER = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
message = input("Enter the message: ")
shift = int(input("Enter the shift amount (positive to encode, negative to decode): "))

encoded_message = ""
i = 0
length = len(message)
while i < length:
    char = message[i]

    if char in LOWER:
        current_index = LOWER.index(char)
        new_index = (current_index + shift) % 26
        encoded_message += LOWER[new_index]

    elif char in UPPER:
        current_index = UPPER.index(char)
        new_index = (current_index + shift) % 26
        encoded_message += UPPER[new_index]

    else:

        encoded_message += char

    i += 1
print(f"Encoded message: {encoded_message}")
```
