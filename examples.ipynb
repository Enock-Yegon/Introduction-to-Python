```python
#price of a meal
meal = input("What's is your order:")
price = float(input(f" Enter the price of your {meal} "))
tax = price * 0.16
tip = price * 0.18
total = tax + tip + price
print(f"tax amount {tax}")
print(f"tip amount {tip}")
print(f"The grand total for the meal is {total:.2f} inclusive of the tax and the tip")
```

```python
# Grade Converter

# Get input and automatically convert it to uppercase
grade = input("Enter your grade (A-F): ").upper()

# Check the grade
if grade == "A":
    print("You scored above 80")
elif grade == "B":
    print("You scored between 71 and 80")
elif grade == "C":
    print("You scored between 61 and 70")
elif grade == "D":
    print("You scored between 51 and 60")
elif grade == "E":
    print("You scored between 41 and 50")
elif grade == "F":
    print("You scored below 40")
else:
    print("The grade entered is invalid")
```
