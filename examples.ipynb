```python
#price of a meal
meal = input("What's is your order:")
price = float(input(f" Enter the price of your {meal} "))
tax = price * 0.16
tip = price * 0.18
total = tax + tip + price
print(f"tax amount {tax}")
print(f"tip amount {tip}")
print(f"The grand total for the meal is {total} inclusive of the tax and the tip")
```
