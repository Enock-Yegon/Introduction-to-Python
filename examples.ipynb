```python
# Exercise 35: Dog Years
#
# It is commonly said that one human year is equivalent to 7 dog years.
# However this simple conversion fails to recognize that dogs reach adulthood
# in approximately two years. As a result, some people believe that it is
# better to count each of the first two human years as 10.5 dog years, and
# then count each additional human year as 4 dog years.
#
# Write a program that implements the conversion from human years to dog years
# described in the previous paragraph. Ensure that your program works correctly
# for conversions of less than two human years and for conversions of two or
# more human years. Your program should display an appropriate error message
# if the user enters a negative number.

# 1. Read input from the user (using float to allow partial/fractional years)
human_years = float(input("Enter the number of human years: "))

# 2. Check for invalid negative input first
if human_years < 0:
    print("Error: Human years cannot be a negative number.")

# 3. Calculate dog years if the input is valid
else:
    if human_years <= 2:
        # The first two years count as 10.5 dog years each
        dog_years = human_years * 10.5
    else:
        # First 2 years = 21 dog years (2 * 10.5)
        # Any additional years count as 4 dog years each
        dog_years = 21 + (human_years - 2) * 4

    # Display the result formatted to 2 decimal places
    print(f"Equivalent age in dog years: {dog_years:.2f}")
```

```python
# Grade Converter

# Get input and automatically convert it to uppercase
grade = input("Enter your grade (A-F): ").upper()

# Check the grade
if grade == "A":
    print("You scored above 80")
elif grade == "B":
    print("You scored between 71 and 80")
elif grade == "C":
    print("You scored between 61 and 70")
elif grade == "D":
    print("You scored between 51 and 60")
elif grade == "E":
    print("You scored between 41 and 50")
elif grade == "F":
    print("You scored below 40")
else:
    print("The grade entered is invalid")
```
