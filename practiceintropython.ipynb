```python
sale_1 = {"items":"Laptops", "quantity":15, "price":65000}
sale_2 = {"items":"Televisions", "quantity":20, "price":80000}
sale_3 = {"items":"Mouses", "quantity":15, "price":1000}
daily_transactions = [sale_1, sale_2 , sale_3]
print(daily_transactions)
```

```python
total_earning = 0
for transaction in daily_transactions:
    total_earning = total_earning + (transaction["quantity"]["price"]) * transaction
print(f"The total revenue for earned for the 3 days is = {total_earning}")
```

```python

# Our updated dataset including the day and the unit price
daily_transactions = [
    {"day": "Monday", "item": "Notebooks", "quantity": 5, "price": 150},
    {"day": "Tuesday", "item": "Textbooks", "quantity": 2, "price": 1200},
    {"day": "Wednesday", "item": "Pens", "quantity": 10, "price": 30}
]

# Loop through each transaction record
total_earning = 0
for transaction in daily_transactions:
    # 1. Calculate the earnings for this specific transaction (Quantity * Price)
    transaction_total = transaction["quantity"] * transaction["price"]

    # 2. Extract the day name
    day_name = transaction["day"]

    # 3. Output the daily breakdown directly
    print(f"Total earnings for {day_name}: KES {transaction_total}")

    total_earning = total_earning + transaction_total

print(f"Total = {total_earning}")
```

```python
#What were our total earnings, but only for days where we made more than KES 500?
daily_transactions = [
    {"day": "Monday", "item": "Notebooks", "quantity": 5, "price": 150},      # Total: 750
    {"day": "Tuesday", "item": "Textbooks", "quantity": 2, "price": 1200},    # Total: 2400
    {"day": "Wednesday", "item": "Pens", "quantity": 10, "price": 30}         # Total: 300
]

total_earning = 0

for transaction in daily_transactions:
    transaction_total = transaction["quantity"] * transaction["price"]
    day_name = transaction["day"]

    # 1. Check if the day's earning is greater than 500
    if transaction_total > 500:
        print(f"🔥 High earning day! {day_name}: KES {transaction_total}")

        # 2. Only add to the total if the condition is met
        total_earning += transaction_total
    else:
        print(f"Normal day. {day_name}: KES {transaction_total}")

# --- Loop ends ---

print(f"\nTotal (High Earning Days Only) = {total_earning}") #\n - new line keyword
```
