```python
#Develop a program that reads a four-digit integer from the user and displays the sum
#of the digits in the number. For example, if the user enters 3141 then your program
#should display 3+1+4+1=9.
num = input("Enter a 4 digit number of your choice:")
digit_sum = 0
for char in num:
        digit_sum += int(char)

print(f"The sum of the digits in {num} is: {digit_sum}")
```

```python
#Create a program that reads three integers from the user and displays them in sorted
#order (from smallest to largest). Use the min and max functions to find the smallest
#and largest values. The middle value can be found by computing the sum of all three
#values, and then subtracting the minimum value and the maximum value

num_1 = int(input("Enter the first number:"))
num_2 = int(input("Enter the second number:"))
num_3 = int(input("Enter the third number:"))
if num_1 >= num_2 and num_1 >= num_3:
    largest = num_1
elif num_2 >= num_1 and num_2 >= num_3:
    largest = num_2
else:
    largest = num_3
    print(f"The largest number is:{largest}")
```

```python
# 1. Read three integers from the user
# use of min and max funtion to sort from the largest to the smallest
num1 = int(input("Enter first integer: "))
num2 = int(input("Enter second integer: "))
num3 = int(input("Enter third integer: "))

# 2. Find the smallest value using min()
smallest = min(num1, num2, num3)

# 3. Find the largest value using max()
largest = max(num1, num2, num3)

# 4. Find the middle value using arithmetic:
#    (Total Sum) - Smallest - Largest = Middle
print(f"The smallest value is: {smallest}")
print(f"The largest value is: {largest}")
total_sum = num1 + num2 + num3
middle = total_sum - smallest - largest

# 5. Display the numbers in sorted order
print(f"Sorted order: {smallest}, {middle}, {largest}")
```

Creates a list containing the values 10, 20, 30.

Prints the list and calculates the average score.


```python
num = [10,20,30]
num.append(40)
num.insert(1,15)
smallest = min(num)
largest = max(num)
average = sum(num)/len(num)
index = num.index(15)
print(f"The appended list is:{num}")
print(f"The the new list after inserting 15 at index 1 is :{num}")
print(f"The minimum value in the list is:{smallest}")
print(f"The maximum value in the list is:{largest}")
print(f"The average of the list is:{average}")
print(f"The index in 15 is {index}")
```

Create a program that keeps asking the user for numbers until they press =.

The program should then output:

sum

lowest val

highest val

average val

The program should not crash if the user enters a non-numerical val.


```python
numbers = []
while True:
    user_input = input("Enter a number (press = to finish):")
    if user_input == "=":
        break
    elif user_input.isdigit():
         numbers.append(int(user_input))
    else:
        print("Not a number! Try again.")
if numbers:
    sum_of_numbers = sum(numbers)
    lowest_val = min(numbers)
    highest_val = max(numbers)
    average_val = sum_of_numbers / len(numbers)
    print(f"The numbers entered are: {numbers}")
    print(f"Sum: {sum_of_numbers}")
    print(f"Lowest Value: {lowest_val}")
    print(f"Highest Value: {highest_val}")
    print(f"Average Value: {average_val}")
else:
    print("Enter a correct number.")
```
