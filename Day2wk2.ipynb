EXAMPLE 108(The Python Workbook)
Create a program that reads integers from the user until a blank line is entered. Once
all of the integers have been read your program should display all of the negative
numbers, followed by all of the zeros, followed by all of the positive numbers. Within
each group the numbers should be displayed in the same order that they were entered by the user. For example, if the user enters the values 3, -4, 1, 0, -1, 0, and -2 then
your program should output the values -4, -1, -2, 0, 0, 3, and 1. Your program
should display each value on its own line.

```python
# Solution 1
numbers = []
while True:
  user_input = input("Enter any integers of your choice(when done enter a blank line):")
  if user_input == " ":
    break
  elif user_input.lstrip("-").isdigit():
    numbers.append(int(user_input))
  else:
    print("Invalid input. Please enter a valid integer.")
negatives = [num for num in numbers if num < 0]
zeros = [num for num in numbers if num == 0]
positives = [num for num in numbers if num > 0]
print(f"\nThe negatives are {negatives}")
print(f"The zeros are {zeros}")
print(f"The positives are {positives}")
print(f"\n{negatives},{zeros},{positives}")


```

EXAMPLE 107
In this exercise, you will create a program that reads words from the user until the
user enters a blank line. After the user enters a blank line your program should display each word entered by the user exactly once. The words should be displayed in
the same order that they were entered. For example, if the user enters:
first
second
first
third
second
then your program should display:
first
second
third


```python
words = []
while True:
  user_input = input("Enter any words of your choice(when done enter a blank line):")
  if user_input == " ":
    break
  elif user_input not in words:
    words.append(user_input)
for word in words:
  print(f"{word}")
```

Build a complete Student Record program that:

Stores student details inside a structured Dictionary.
Implements a list comprehension to extract failing scores.
Uses a loop to display and filter multiple nested records
cleanly.


```python
students = {}
while True:
    name = input("Enter student name (press Enter to stop): ")
    if name == " ":
        break

    subject1 = input("Enter subject 1: ")
    subject2 = input("Enter subject 2: ")
    subject3 = input("Enter subject 3: ")

    score1 = int(input(f"Enter score for {subject1} : "))
    score2 = int(input(f"Enter Score for {subject2}: "))
    score3 = int(input(f"Enter Score for {subject3}: "))

    students[name] = {
        "subjects": [subject1, subject2, subject3],
        "scores": [score1, score2, score3]
    }

for name, details in students.items():
    print(f"\nStudent: {name}")
    print(f"Subject:  {details['subjects']}")
    print(f"Scores:  {details['scores']}")

    failing = [score for score in details["scores"] if score < 50]

    if failing:
        print(f"Failing Scores: {failing}")
    else:
        print("✅ Passed everything!")
```
