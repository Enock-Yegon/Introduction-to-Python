```python
#Mini calculator
num1 = float(input("Enter the first Number: "))
num2 = float(input("Enter the second Number: "))
print(f"The sum of the two numbers is {num1 + num2}")
print(f"The difference between the two numbers is {num1 - num2}")
print(f"The product of the two numbers is {num1 * num2}")
print(f"Num1 is more than num2 by {num1 / num2}")
print(f"The modulus of the number is {num1 % num2}")
print(f"The floor of the division of the two numbers is {num1 // num2}")
print(f"The power of num1 to num2 is {num1 ** num2}")
```

```python
width = float(input("Enter the width of the room "))
lengh = float(input(" Enter the length of the room "))
area = lengh * width
print(f" The area of the room in square metres is {area}")
```
