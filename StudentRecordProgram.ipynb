#### Student Record Program
- stores details inside structured dictionary
- implement list comprehension
- add detail
- record marks
- views failings
- view complete details
- exit


`newlist = [expression for item in iterable if condition == True]`

```python

scoresys = {}

while True:

    # Menu
    print("==============================")
    print("   STUDENT RECORD PROGRAM")
    print("==============================")
    print("1. Add Student")
    print("2. Record score")
    print("3. View Failings")
    print("4. View Complete Details")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        score = int(input("Enter first mark: "))
        # store the student details in nested dictionary
        # store score inside a list
        scoresys[name] = {
            "score": [score]
        }
        print("Student added successfully.")

    elif choice == "2":
        #editing values in dictionary
        student_to_edit = input("Enter student name to edit: ")
        if student_to_edit in scoresys: # checking if the student exists in the dictionary using the in operator
            scoresys[student_to_edit] = {"score": [int(input("Enter new marks for " + student_to_edit + ": "))]}
            for name in scoresys:
                print(name, "\t\t", scoresys[name]["score"])
            else:
              print("Student not found.")

    elif choice == "3":
        print("\nStudents who failed (Average below 50)")
        print("---------------------------------------")

       # Used list comprehension to create a list of students who failed
        failscore = [fail for fail in scoresys if scoresys[fail]["score"][0] < 50]
      # iterating through the list of students who failed and printing their names and scores
        for student in failscore:
          print(student,"\t\t", scoresys[student]["score"])
        else:
            print("No failing students.")

    elif choice == "4":
        # check is dict is not empty
        if scoresys:

            print("==============================================")
            print("Student         score            Average")
            print("==============================================")
            for student in scoresys:
                # display the student details
                print(student,"\t\t",scoresys[student]["score"])

        else:
            print("No student records available.")
    elif choice == "5":

        print("EXIT!!")
        break
    else:

        # Runs if the user enters anything other than 1-5
        print("Invalid choice. Please try again.")
```
