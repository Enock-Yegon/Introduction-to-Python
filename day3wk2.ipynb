## Exercise 47: Birth Date to Astrological Sign

### Overview
Astrological horoscopes divide the year into twelve 30-day (approximate) periods corresponding to specific zodiac signs based on the Sun's position.

This program takes a user's **month** and **day of birth**, validates that the entered date actually exists on the calendar, and outputs their corresponding zodiac sign.

---

### Zodiac Sign Boundaries

| Zodiac Sign | Date Range |
| :--- | :--- |
| **Capricorn** | December 22 – January 19 |
| **Aquarius** | January 20 – February 18 |
| **Pisces** | February 19 – March 20 |
| **Aries** | March 21 – April 19 |
| **Taurus** | April 20 – May 20 |
| **Gemini** | May 21 – June 20 |
| **Cancer** | June 21 – July 22 |
| **Leo** | July 23 – August 22 |
| **Virgo** | August 23 – September 22 |
| **Libra** | September 23 – October 22 |
| **Scorpio** | October 23 – November 21 |
| **Sagittarius** | November 22 – December 21 |

---

### Implementation Features
1. **Flexible Inputs:** Accepts month names (e.g., `"March"`, `"march"`) or numbers (`1`–`12`).
2. **Boundary Validation:** Verifies that the entered day falls within the maximum valid days for that month (e.g., prevents entering `Feb 30` or `March 57`).
3. **Compound Conditionals:** Uses crisp `if / elif` logic to match the birth window.




```python
# 1. Ask user for input
month = input("Enter your birth month (e.g. March or 3): ").strip().lower()
day = int(input("Enter your birth day (e.g. 21): ").strip())

# Convert numeric month entry to month name if needed
months_list = [
    "january", "february", "march", "april", "may", "june",
    "july", "august", "september", "october", "november", "december"
]

if month.isdigit():
    month_num = int(month)
    if 1 <= month_num <= 12:
        month = months_list[month_num - 1]

# Days in each month dictionary
max_days = {
    "january": 31, "february": 29, "march": 31, "april": 30,
    "may": 31, "june": 30, "july": 31, "august": 31,
    "september": 30, "october": 31, "november": 30, "december": 31
}

# 2. VALIDATION CHECK: Is the day valid for this month?
if month not in max_days or day < 1 or day > max_days[month]:
    print(f"\nInvalid date! {month.capitalize()} does not have {day} days.")
else:
    # 3. Determine Zodiac Sign
    sign = ""

    if (month == "december" and day >= 22) or (month == "january" and day <= 19):
        sign = "Capricorn"
    elif (month == "january" and day >= 20) or (month == "february" and day <= 18):
        sign = "Aquarius"
    elif (month == "february" and day >= 19) or (month == "march" and day <= 20):
        sign = "Pisces"
    elif (month == "march" and day >= 21) or (month == "april" and day <= 19):
        sign = "Aries"
    elif (month == "april" and day >= 20) or (month == "may" and day <= 20):
        sign = "Taurus"
    elif (month == "may" and day >= 21) or (month == "june" and day <= 20):
        sign = "Gemini"
    elif (month == "june" and day >= 21) or (month == "july" and day <= 22):
        sign = "Cancer"
    elif (month == "july" and day >= 23) or (month == "august" and day <= 22):
        sign = "Leo"
    elif (month == "august" and day >= 23) or (month == "september" and day <= 22):
        sign = "Virgo"
    elif (month == "september" and day >= 23) or (month == "october" and day <= 22):
        sign = "Libra"
    elif (month == "october" and day >= 23) or (month == "november" and day <= 21):
        sign = "Scorpio"
    elif (month == "november" and day >= 22) or (month == "december" and day <= 21):
        sign = "Sagittarius"

    print(f"\nBorn on {month.capitalize()} {day}, your zodiac sign is {sign}.")
```

## Exercise 48: Chinese Zodiac Calculator

### Overview
The Chinese Zodiac assigns an animal to each year in a repeating 12-year cycle. Because the cycle repeats predictably every 12 years, we can determine the corresponding animal for **any year** using the **modulo operator (`%`)**.

---

### The 12-Year Cycle Standard
| Remainder (`year % 12`) | Animal | Reference Year |
| :--- | :--- | :--- |
| **8** | Dragon | 2000 |
| **9** | Snake | 2001 |
| **10** | Horse | 2002 |
| **11** | Sheep | 2003 |
| **0** | Monkey | 2004 |
| **1** | Rooster | 2005 |
| **2** | Dog | 2006 |
| **3** | Pig | 2007 |
| **4** | Rat | 2008 |
| **5** | Ox | 2009 |
| **6** | Tiger | 2010 |
| **7** | Hare | 2011 |

---

### Implementation Strategy
1. Prompt the user for a valid year ($\ge 0$).
2. Compute `year % 12` to find the offset within the 12-year array.
3. Map the resulting index directly to the animal list and display the result.


```python
#Solution 1
animals = [
    "Monkey",
    "Rooster",
    "Dog",
    "Pig",
    "Rat",
    "Ox",
    "Tiger",
    "Hare",
    "Dragon",
    "Snake",
    "Horse",
    "Sheep",
]
year = int(input("Enter a year (>= 0): "))
if year < 0:
    print("Please enter a year greater than or equal to 0.")
else:
    animal = animals[year % 12]
    print(f"\nThe animal associated with the year {year} is the {animal}.")
```
